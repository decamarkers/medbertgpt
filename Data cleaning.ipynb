```python
import pandas as pd
webmd = pd.read_json('webmdQAs.json')
ehealth = pd.read_json('ehealthforumQAs.json')
iclinic = pd.read_json('icliniqQAs.json')
questionDoctor = pd.read_json('questionDoctorQAs.json')
```

```python
frames = [webmd[["answer","question","url","tags"]],ehealth[["answer","question","url","tags"]],iclinic[["answer","question","url","tags"]],questionDoctor[["answer","question","url","tags"]]]
result = pd.concat(frames, axis=0, ignore_index=True)
```

```python
result
```

```python
result.to_excel("medicalQnA.xlsx")
```

```python
result.dropna(subset=['question', 'answer'], inplace=True)
```

```python
result.drop_duplicates(subset=['question', 'answer'], inplace=True)
```

```python
import re

# Define a function to process the text
def process_text(text):
    # Convert to lowercase and keep only allowed special characters
    processed_text = re.sub(r'[^a-z0-9\s\-/]', '', text.lower())
    return processed_text

# Apply the function to the 'question' and 'answer' columns
result['question'] = result['question'].apply(process_text)
result['answer'] = result['answer'].apply(process_text)
```

```python
result
```

```python
# Assuming you have the 'nltk' library installed
import nltk
nltk.download('punkt')

result['question'] = result['question'].apply(nltk.word_tokenize)
result['answer'] = result['answer'].apply(nltk.word_tokenize)
```

```python
# Assuming you have the 'nltk' library installed
from nltk.corpus import stopwords
nltk.download('stopwords')

stop_words = set(stopwords.words('english'))

result['question'] = result['question'].apply(lambda tokens: [word for word in tokens if word not in stop_words])
result['answer'] = result['answer'].apply(lambda tokens: [word for word in tokens if word not in stop_words])
```

```python
# Assuming you have the 'nltk' library installed
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')

lemmatizer = WordNetLemmatizer()

result['question'] = result['question'].apply(lambda tokens: [lemmatizer.lemmatize(word) for word in tokens])
result['answer'] = result['answer'].apply(lambda tokens: [lemmatizer.lemmatize(word) for word in tokens])
```

```python
result['question'] = result['question'].apply(' '.join)
result['answer'] = result['answer'].apply(' '.join)
```
